```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(['Unnamed: 25','User'],axis=1)
```

```python
df.info()
```

```python
df['Category 12']=df['Category 12'].fillna(df['Category 12'].mean())
```

```python
df['Category 24']=df['Category 24'].replace(np.nan,df['Category 24'].mean())
```

```python
df['Category 11']=df['Category 11'].replace('2\t2.',0)
```

```python
df['Category 11'].unique()
```

```python
df['Category 11']=df['Category 11'].astype(float)
```

```python
df.columns
```

```python
df.rename(columns={'Category 1':'Churches'},inplace=True)
```

```python
df.rename(columns={'Category 2':'resorts','Category 3':'Beaches','Category 4':'Parks','Category 5':'Theatre'},inplace=True)
```

```python
df.rename(columns={'Category 6':'museums','Category 7':'mall','Category 8':'zoo','Category 9':'restaurant','Category 10':'pubs/bars'},inplace=True)
```

```python
df.rename(columns={'Category 11':'Localservice','Category 12':'burger/pizza-shop','Category 13':'Hotels/lodges',
                   'Category 14':'Juicebars','Category 15':'art_gallery'},inplace=True)
```

```python
df.rename(columns={'Category 16':'danceclub','Category 17':'swimmimgpool','Category 18':'gym',
                   'Category 19':'bakeries','Category 20':'beauty&spa'},inplace=True)
```

```python
df.rename(columns={'Category 21':'cafe','Category 22':'viewpoints','Category 23':'monuments',
                   'Category 24':'gardens'},inplace=True)
```

```python
n_df=df
```

```python
from scipy.stats import zscore
```

```python
df_scaled=n_df.apply(zscore)
df_scaled.head(10)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=2, affinity='euclidean',  linkage='average')
```

```python
model.fit(df_scaled)
```

```python
n_df['labels'] = model.labels_
n_df.head(10)
```

```python
from scipy.cluster.hierarchy import cophenet,dendrogram, linkage
```

```python
from scipy.spatial.distance import pdist 
```

```python

Z = linkage(df_scaled, metric='euclidean', method='average')
c, coph_dists = cophenet(Z , pdist(df_scaled))

c
```

```python
plt.figure(figsize=(10, 5))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z, leaf_rotation=90.,color_threshold = 40, leaf_font_size=8. )
plt.tight_layout()
```

```python
Z = linkage(df_scaled, metric='euclidean', method='ward')
c, coph_dists = cophenet(Z , pdist(df_scaled))

c
```

```python
plt.figure(figsize=(10, 5))
plt.title('Agglomerative Hierarchical Clustering Dendogram')
plt.xlabel('sample index')
plt.ylabel('Distance')
dendrogram(Z, leaf_rotation=90.,color_threshold=600,  leaf_font_size=10. )
plt.tight_layout()
```
